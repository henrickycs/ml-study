```python
y = abt['Total Charges']
X = abt.drop('Total Charges', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state = 7)
```

```python
print(len(X_train), len(X_test),len(y_train),len(y_test))
```

```python
#preprocessing
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#import models
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
```

```python
pipelines = {
    'rf':make_pipeline(StandardScaler(), RandomForestRegressor(random_state = 7)),
    'gb':make_pipeline(StandardScaler(), GradientBoostingRegressor(random_state = 7)),
    'lasso':make_pipeline(StandardScaler(), Lasso(random_state = 7)),
    'ridge':make_pipeline(StandardScaler(), Ridge(random_state = 7)),
    'enet':make_pipeline(StandardScaler(), ElasticNet(random_state = 7))
}
```

```python
grid = {
    'rf':{'randomforestregressor__n_estimators':[10,20,30]},
    'gb':{'gradientboostingregressor__alpha':[0.5,0.9,0.99]},
    'lasso':{'lasso__alpha':[0.5,0.9,0.99]},
    'ridge':{'ridge__alpha':[0.5,0.9,0.99]},
    'enet':{'elasticnet__alpha':[0.5,0.9,0.99]}
}
```

```python
RandomForestRegressor().get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
fit_models = {}
for algo, pipeline in pipelines.items():
    model = GridSearchCV(pipeline, grid[algo], cv = 10, n_jobs =-1)
    model.fit(X_train, y_train)
    fit_models[algo] = model
    print(algo, 'model has been fit')
```

```python
fit_models
```

```python
fit_models['rf']
```

```python
fit_models['rf'].predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for algo, model in fit_models.items():
    yhat = model.predict(X_test)
    print('MSE for', algo, 'is', mean_squared_error(y_test, yhat))
    
#Best model --> RandomForestRegressor
```

```python
bestmodel = fit_models['rf'].best_estimator_
```

```python
bestmodel.predict(X_test)
```
